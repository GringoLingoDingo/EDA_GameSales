```python
import requests

# Step 1: Get the full list of Steam apps
app_list_url = "http://api.steampowered.com/ISteamApps/GetAppList/v2/"
response = requests.get(app_list_url)
data = response.json()

# Check the structure and extract the apps list
apps = data.get("applist", {}).get("apps", [])
print(f"Total number of apps fetched: {len(apps)}")

# Create a list of app IDs (if you need names as well, you can store them as tuples)
app_ids = [app["appid"] for app in apps]
```

```python
import requests
import time

def get_game_name(appid):
    """Retrieve the game name using the appdetails API."""
    url = f"https://store.steampowered.com/api/appdetails?appids={appid}"
    try:
        data = requests.get(url).json()
        # The response is keyed by the appid as a string.
        if data.get(str(appid)) and data[str(appid)].get("success"):
            return data[str(appid)]["data"]["name"]
    except Exception as e:
        print(f"Error fetching game name for appid {appid}: {e}")
    return None

def get_review_summary(appid):
    """Retrieve the review summary (positive and negative counts) using the appreviews API."""
    review_url = f"https://store.steampowered.com/appreviews/{appid}?json=1&filter=recent&language=all&purchase_type=all"
    try:
        review_data = requests.get(review_url).json()
        # Check if the query_summary field exists
        if "query_summary" in review_data:
            qs = review_data["query_summary"]
            positive = qs.get("total_positive", 0)
            negative = qs.get("total_negative", 0)
            return positive, negative
    except Exception as e:
        print(f"Error fetching reviews for appid {appid}: {e}")
    return None, None

# Example usage: let's use a single app id for demonstration.
# You mentioned appid "125116" as an example.
appid = app_ids[0:70]
game_name = get_game_name(appid)
positive, negative = get_review_summary(appid)
#
#if game_name is not None:
#    print(f"Game: {game_name}")
#    print(f"Positive Reviews: {positive}")
#    print(f"Negative Reviews: {negative}")
#else:
#    print(f"No details found for appid {appid}.")
```

```python
Steam_reviews = {"Title":[],
                 "Positive Reviews":[],
                 "Negative Reviews":[]
                  ,"AppID":[]}
```

```python
for appid in app_ids:
    game_name = get_game_name(appid)
    positive, negative = get_review_summary(appid)
    
    if game_name is not None:
        Steam_reviews["Title"].append(game_name)
        Steam_reviews["Positive Reviews"].append(positive)
        Steam_reviews["Negative Reviews"].append(negative)
        Steam_reviews["AppID"].append(appid)
    else:
        print(f"No details found for appid {appid}.\n")
    
    # Optional: add a delay to avoid overwhelming the API
    time.sleep(0.8)
```
